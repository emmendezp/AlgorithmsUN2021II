#1. Fibonacci Number

```python
def calc_fib(n):
    if (n <= 1):
        return n

    return calc_fib(n - 1) + calc_fib(n - 2)

def fib_n1(n):
    x = 0
    y = 1

    if n == 0:
      return 0
    if n == 1:
      return 1

    for i in range(n-1):
        z = x + y
        x = y
        y = z
  
    return z


n = int(input())
print(fib_n1(n))
```

#2. Last_digit_of_fibonacci_number

```python
def get_fibonacci_last_digit_naive(n):
    if n <= 1:
        return n

    previous = 0
    current  = 1

    for _ in range(n - 1):
        previous, current = current, previous + current

    return current % 10

def fib_n2(n):

    if n == 0:
      return 0
    if n == 1:
      return 1

    fib_list = [0, 1]
    for i in range(n-1):
        fib_list.append(fib_list[-1]%10 + fib_list[-2]%10)

    return fib_list[n]%10

#if __name__ == '__main__':
    #input = sys.stdin.read()
n = int(input())
print(fib_n2(n))
```

#3. Greatest_common_divisor

```python
def gcd_naive(a, b):
    current_gcd = 1
    for d in range(2, min(a, b) + 1):
        if a % d == 0 and b % d == 0:
            if d > current_gcd:
                current_gcd = d

    return current_gcd

def get_gcd(a, b):
    min_val = min(a, b)
    max_val = max(a, b)

    if min_val == 0:
        return max_val
    elif min_val == 1:
        return 1
    else:
        return get_gcd(min_val, max_val % min_val)

#if __name__ == "__main__":
   # input = sys.stdin.read()
a, b = map(int, input().split())
print(get_gcd(a, b))
```

#4. least_common_multiple

```python
def lcm_naive(a, b):
    for l in range(1, a*b + 1):
        if l % a == 0 and l % b == 0:
            return l

    return a*b

def gcd(a, b):
    return a if not b else gcd(b, a % b)
 
def lcm(a, b):
    return a * b / gcd(a, b)

#if __name__ == '__main__':
    #input = sbs.stdin.read()
a, b = map(int, input().split())
print(int(lcm(a, b)))
```

#5. fibonacci_number_again

```python
def get_pisano_period(m):
  a = 0
  b = 1
  c = a + b

  for i in range(m * m):
    c = (a + b) % m
    a = b
    b = c
    if (a == 0 and b == 1):
      return i + 1

def get_fib_huge(n, m):
  aux = n % get_pisano_period(m)
  num_1 = 0
  num_2 = 1
  res = aux

  for i in range(aux - 1):
    res = (num_1 + num_2) % m
    num_1 = num_2
    num_2 = res
  
  f_h = res % m
  return f_h

n, m = map(int, input().split())
print(get_fib_huge(n, m))
```

#6. last_digit_of_the_sum_of_fibonacci_numbers

```python
def fib_sum(n):

    f_0 = 0
    f_1 = 1

    if n == 0:
      return 0
    if n == 1:
      return 1
    else:
      rem = n % 60
      if (rem == 0):
        return 0

    for i in range(2, rem + 3):
      f = (f_0 + f_1) % 60
      f_0 = f_1
      f_1 = f
    
    sum_fib = f_1 -1
    
    return sum_fib % 10

n = int(input())
print(fib_sum(n))
```
