# II Solving The Maximum Pairwise Product Programming Challenge in C++

## C++

```python
%%writefile max_pairwise_product1.cpp
#include <iostream>
#include <vector>
#include <algorithm>

int MaxPairwiseProduct(const std::vector<int>& numbers) {
    int max_product = 0;
    int n = numbers.size();

    for (int first = 0; first < n; ++first) {
        for (int second = first + 1; second < n; ++second) {
            max_product = std::max(max_product,
                numbers[first] * numbers[second]);
        }
    }

    return max_product;
}

int main() {
    int n;
    std::cin >> n;
    std::vector<int> numbers(n);
    for (int i = 0; i < n; ++i) {
        std::cin >> numbers[i];
    }

    std::cout << MaxPairwiseProduct(numbers) << "\n";
    return 0;
}
```

```python
%%writefile max_pairwise_product_eficiente.cpp
#include <iostream>
#include <vector>
#include <algorithm>
#include <list>

long long MaxPairwiseProductEficiente(const std::vector<int>& numbers, long long &execints) {
    int n = numbers.size();                                             //1
    execints++;

    long long max_product = 0LL;                                        //1
    execints++;

    long long mayor = 0;                                                //1
    execints++;

    long long smayor = 0;                                               //1
    execints++;

    execints++;
    long long co1 = 0LL;
    long long co2 = 0LL;
    long long co3 = 0LL;  
    for(int i = 0; i < n; i++){                                         //1+2n
        execints+=2;

        
        execints++;                                                     //n
        if(numbers[i]>mayor){
          
          smayor = mayor;                                             //[0,2n]
          execints++;

          mayor = numbers[i];                                         
          execints++;

          co1++;

        }else{                                                       //[0,n]
            co2++;
                                                                
            execints++;
            if(numbers[i]>smayor){   
                co3++;

                smayor = numbers[i];                                  //[0,n]
                execints++;
            }   
        }
        
    }
    //std::cout<<"co1: "<<co1<<"\n";
    //std::cout<<"co2: "<<co2<<"\n";
    //std::cout<<"co3: "<<co3<<"\n";
    max_product = (long long)mayor*smayor;                              //1
    execints++;
                                                              //--------------
    return max_product;                                       //  6 + [3n,8n]
}

int main() {         
    long long execints =0LL;                                              

    int n;                                                              //1
    execints++;

    std::cin >> n;                                                      //1
    execints++;                                                    

    std::vector<int> numbers(n);                                        //n
    execints+=n;   
                                        
    long long maxproduc = 0LL;                                          //1
    execints++;


    execints++;
    for (int i = 0; i < n; ++i) {                                       //1 + 2n
        execints+=2;

        std::cin >> numbers[i];                                         //n
        execints++;

    }

    maxproduc = MaxPairwiseProductEficiente(numbers,execints);          //  6 + [3n,8n]
    execints++;                                                         //1

    std::cout<<maxproduc<<"\n";       
    
    //std::cout <<"execints: "<<execints << "\n";                          //-----------------
    return 0;                                               //    11 + [7n, 12n]
}
```

##Efficient solution

```python
%%writefile max_pairwise_productE.py
def max_pairwise_product(numbers):
    n = len(numbers)
    first = 0 
    second = 0
    for i in range(n):
        if numbers[i] > first:
			      second = first
			      first = numbers[i]
        elif numbers[i] > second:
				      second = numbers[i]
              
    return (first * second)


if __name__ == '__main__':
    input_n = int(input())
    input_numbers = [int(x) for x in input().split()]
    print(max_pairwise_product(input_numbers))
```
